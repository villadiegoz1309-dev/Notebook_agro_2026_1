## ¿Qué es Python?

Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser:

* **Legible:** Es fácil de leer y escribir, ideal para ingenieros que no son especialistas en software.
* **Versátil:** Sirve para analizar datos, automatizar maquinaria, crear aplicaciones web o entrenar Inteligencia Artificial.
* **Comunitario:** Tiene miles de "librerías" (herramientas listas para usar) creadas por otros expertos, lo que te ahorra mucho trabajo.

### 1. Cálculo del Índice de Madurez

Este código ayuda a determinar si un lote de frutas está listo para cosecha basándose en la relación entre sólidos solubles ($^\circ Brix$) y acidez.

```
# Datos de la muestra
grados_brix = 12.5
acidez_titulable = 0.8  # Porcentaje de ácido cítrico

def calcular_indice_madurez(brix, acidez):
    indice = brix / acidez
    return indice

resultado = calcular_indice_madurez(grados_brix, acidez_titulable)
print(f"El índice de madurez es: {resultado:.2f}")

if resultado > 15:
    print("Estado: Óptimo para cosecha/consumo.")
else:
    print("Estado: Aún está verde o requiere más tiempo.")
```

### 2. Estimación de Humedad en Secado

Un paso crítico en agroindustria es el secado de granos o harinas. Este código calcula el contenido de humedad en base húmeda.

```
# Pesos en gramos
peso_inicial = 500.0  # Muestra fresca
peso_seco = 410.0     # Después de la estufa

def porcentaje_humedad(inicial, seco):
    humedad = ((inicial - seco) / inicial) * 100
    return humedad

h_final = porcentaje_humedad(peso_inicial, peso_seco)
print(f"El contenido de humedad es del {h_final}%")
```

### 3. Monitoreo de Temperatura en Almacén

Simulación de un sistema que alerta si un silo o cuarto frío se sale del rango permitido.

```
# Lista de temperaturas registradas por un sensor en una hora
temperaturas = [4.2, 4.5, 5.1, 6.8, 7.2, 5.5]

limite_maximo = 6.0

print("Analizando registros térmicos...")
for t in temperaturas:
    if t > limite_maximo:
        print(f"ALERTA: Temperatura de {t}°C excede el límite permitido!")
    else:
        print(f"Temperatura {t}°C: OK")
```
